```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from scipy.cluster.vq import whiten
from collections import Counter


# read data
df = pd.read_csv('data/treesna_tree_data.csv')
# use coordinates
features = np.array(df.iloc[:, [1,2]])
# normalize features
features = whiten(features)

# fit k-means
# specify k=8 (8 clusters)
cls = KMeans(n_clusters=8)
cls.fit(features)



# cluster centers
centers = pd.DataFrame(cls.cluster_centers_, columns=('latitude', 'longitude'))


# count samples in each class
counts = Counter(cls.predict(features))
print('Samples in each class: %s' % counts)


# attach class labels and counts to centers
centers['label'] = counts.keys()
centers['counts'] = counts.values()

centers
```
