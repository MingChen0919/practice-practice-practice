```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV

from sklearn.decomposition import PCA
```

```python
# load data
iris = load_iris()
X = iris.data
y = iris.target

# scale features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# build parameters grid
C_range = np.logspace(-2, 10, 13)
gamma_range = np.logspace(-9, 3, 13)
param_grid = dict(gamma=gamma_range, C=C_range)
# create cross validation instance
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
grid = GridSearchCV(SVC(), param_grid=param_grid, cv=cv)
grid.fit(X, y)

print("The best parameters are %s with a score of %0.2f"
      % (grid.best_params_, grid.best_score_))
```

```python
# visualization
# use PCA to reduce X's dimensions
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)
# convert y to a two classes label
y[y!= 1] = 0

# choose some parameters to fit a model
C_vis = [1e-1, 1, 1e2]
gamma_vis = [1e-2, 1e-1, 1]
classifiers = []
for C in C_vis:
    for gamma in gamma_vis:
        clf = SVC(C=C, gamma=gamma)
        clf.fit(X_reduced, y)
        classifiers.append((C, gamma, clf))
```

```python
# draw visualization of parameter effects
plt.figure(figsize=(10, 8))
xx, yy = np.meshgrid(np.linspace(-4, 4, 200), np.linspace(-4, 4, 200))
for (k, (C, gamma, clf)) in enumerate(classifiers):
    # evaluate decision function in a grid
    Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.subplot(len(C_vis), len(gamma_vis), k+1)
    plt.title("gamma=10^%d, C=10^%d" % (np.log10(gamma), np.log10(C)),
              size='medium')
    plt.pcolormesh(xx, yy, -Z, cmap=plt.cm.RdBu)
    plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, 
                edgecolors='k', cmap=plt.cm.hot)
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    if C == grid.best_params_['C'] and gamma == grid.best_params_['gamma']:
        plt.xlabel("Effect of best parameters, score = %0.2f" % grid.best_score_)
```

```python
scores = grid.cv_results_['mean_test_score'].reshape(len(C_range), len(gamma_range))
plt.figure(figsize=(8,6))
plt.imshow(scores, interpolation='nearest', cmap=plt.cm.hot)
plt.colorbar()
plt.xticks(np.arange(len(gamma_range)), gamma_range, rotation=45)
plt.yticks(np.arange(len(C_range)), C_range)
plt.xlabel('gamma')
plt.ylabel('C')
plt.title('Cross validation score')
plt.show()
```
